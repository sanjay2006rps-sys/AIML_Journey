```python
"""
student_demo.py
----------------
A simple, self-contained demo of a Student class.
"""


class Student:
    """Represents a single student."""

    HONORS_THRESHOLD = 8.5

    def __init__(self, name: str, age: int, department: str, cgpa: float):
        self.name = name
        self.age = age
        self.department = department
        self.cgpa = cgpa

    def display(self) -> None:
        """Print a formatted summary of the student's details."""
        print(
            f"Name: {self.name:<15} Age: {self.age:<3} "
            f"Department: {self.department:<20} CGPA: {self.cgpa}"
        )

    def update_cgpa(self, new_cgpa: float) -> None:
        """Update the student's CGPA after validating the new value."""
        if not (0.0 <= new_cgpa <= 10.0):
            raise ValueError("CGPA must be between 0.0 and 10.0")
        self.cgpa = new_cgpa

    def is_honors(self) -> bool:
        """Return True if the student's CGPA meets the honors threshold."""
        return self.cgpa >= Student.HONORS_THRESHOLD


def main():
    # Create at least three students
    s1 = Student("Aarav Sharma", 21, "Computer Science", 8.2)
    s2 = Student("Priya Nair", 20, "Electronics", 9.1)
    s3 = Student("Karthik Raja", 22, "Mechanical", 7.6)

    students = [s1, s2, s3]

    # Display all details
    print("All student details:")
    for s in students:
        s.display()

    # Update one student's CGPA
    print(f"\nUpdating {s1.name}'s CGPA from {s1.cgpa} to 8.9 ...")
    s1.update_cgpa(8.9)

    print("\nDetails after update:")
    s1.display()

    # Print honors status for each student
    print("\nHonors status:")
    for s in students:
        status = "Honors" if s.is_honors() else "Not Honors"
        print(f"{s.name}: {status} (CGPA: {s.cgpa})")


if __name__ == "__main__":
    main()
```
